```python
def common_ancestor(li):
    a_count = []
    c_count = []
    g_count = []
    t_count = []


    for col in range(len(li[0])):
        a = 0
        c = 0
        g = 0
        t = 0
        #print(" ")
        for row in li:
            for nucleotide in row[col]:
                if nucleotide == "A":
                    a +=1
                elif nucleotide == "C":
                    c +=1
                elif nucleotide == "G":
                    g +=1
                elif nucleotide == "T":
                    t +=1 
        a_count.append(a)
        c_count.append(c)
        t_count.append(t)
        g_count.append(g)

    ancestor_string = ""    

    for i in range(len(a_count)):
        if a_count[i] == (max(a_count[i],c_count[i],g_count[i],t_count[i])):
            ancestor_string += "A"
        elif c_count[i] == (max(a_count[i],c_count[i],g_count[i],t_count[i])):
            ancestor_string += "C"
        elif g_count[i] == (max(a_count[i],c_count[i],g_count[i],t_count[i])):
            ancestor_string += "G"
        elif t_count[i] == (max(a_count[i],c_count[i],g_count[i],t_count[i])):
            ancestor_string += "T"
    
    a_str = "A: "
    c_str = "C: "
    g_str = "G: "
    t_str = "T: "
    
    for i in range(len(li[0])):
        a_str += str(a_count[i]) + " "
        c_str += str(c_count[i]) + " "
        g_str += str(g_count[i]) + " "
        t_str += str(t_count[i]) + " "
        

    return ancestor_string, a_str, c_str, g_str, t_str
        
```
